Модель не сможет учиться, если целевая переменная хранится как текст.

````python
df['weight'] = df['weight'].str.replace(',', '.', regex=False).astype(float)
````

6.2. Распределение целевой переменной

Максимальное значение массы значительно превышает медиану, что также указывает на наличие правосторонней асимметрии распределения.

````python
plt.figure(figsize=(8, 5))
sns.histplot(df['weight'], kde=True)
plt.title('Распределение массы черепах')
plt.xlabel('Масса, кг')
plt.ylabel('Количество наблюдений')
plt.show()
````

<br/>
<div class="alerфt alert-info"> Гистограмма распределения массы черепах показывает, что распределение целевой переменной не является симметричным.
 
Основная часть наблюдений сосредоточена в диапазоне примерно от **20 до 100 кг**, что соответствует наиболее типичным значениям массы черепах в датасете.

При этом распределение имеет **выраженную правостороннюю асимметрию** (right-skewed distribution): наблюдается длинный правый хвост, который указывает на наличие относительно небольшого числа черепах с очень большой массой.

Также на графике можно заметить **потенциальные выбросы** — значения массы, значительно превышающие основную массу наблюдений (вплоть до 600 кг). Такие значения могут быть связаны как с особенностями данных, так и с возможными ошибками измерений.

Несмотря на наличие асимметрии и возможных выбросов, распределение не препятствует дальнейшему анализу данных и построению модели. Однако при последующей работе с данными необходимо учитывать наличие аномальных значений.</div>

6.3 Корреляция числовых признаков 

Сначала создадим копию только с числовыми колонками:


````python
numeric_df = df.select_dtypes(include=['number'])
````

Теперь корреляция:

````python
plt.figure(figsize=(14, 10))

sns.heatmap(
    numeric_df.corr(),
    cmap='coolwarm',
    center=0,
    annot=True,        # показываем значения корреляции
    fmt=".2f",         # формат чисел
    square=True
)

plt.title('Корреляция числовых признаков')

plt.show()
````

<br/>
<div class="alerфt alert-info"> Тепловая карта корреляций позволяет оценить силу линейной зависимости между числовыми признаками.

Наиболее заметно, что признаки, связанные с размерами тела черепах (длина и ширина ласт, размеры головы и панциря), имеют достаточно высокую положительную корреляцию друг с другом. Это ожидаемо, поскольку все эти параметры отражают общий размер животного.

Особенно важно рассмотреть корреляцию признаков с целевой переменной `weight`. Можно заметить, что наибольшая положительная корреляция наблюдается у следующих признаков:

- длина ласт (`flipper_length_1`, `flipper_length_2`, `flipper_length_3`, `flipper_length_4`)
- ширина ласт (`flipper_width_*`)
- размеры головы (`head_length`, `head_width`)
- размеры панциря (`shell_length`, `shell_width`)

Это логично с биологической точки зрения: масса животного напрямую связана с его габаритами.

Также видно, что технические признаки `id` и `timestamp` практически не имеют корреляции с целевой переменной, поэтому они не несут полезной информации для модели и могут быть исключены из анализа.

Таким образом, можно сделать вывод, что биометрические измерения черепах являются информативными признаками для предсказания их массы.</div>

## Поиск и исправление ошибок умножениея на ×10

В дополнительном сообщении заказчик сообщил, что при внесении данных некоторые измерения длины могли быть ошибочно умножены на 10. Это произошло из-за того, что один из волонтёров предположил, что значения записаны в сантиметрах, и дополнительно умножил их на 10, хотя все измерения должны быть указаны в миллиметрах.

Такие ошибки могут привести к появлению аномально больших значений, которые будут искажать статистику признаков и ухудшать качество модели линейной регрессии. Поэтому необходимо проверить размерные признаки и при необходимости скорректировать такие значения.

На этапе исследовательского анализа данных ранее были построены boxplot-диаграммы для размерных признаков. Эти графики позволили визуально выявить наличие аномально больших значений.

Наиболее заметные выбросы наблюдаются в признаках:

shell_length

shell_width

flipper_length_1

flipper_length_2

flipper_length_3

flipper_length_4

В этих признаках присутствуют значения, значительно превышающие основной диапазон наблюдений. Это может свидетельствовать о возможной ошибке масштабирования (умножении значений на 10).

Таким образом, визуальный анализ указывает на необходимость дополнительной проверки размерных признаков.

**7.2 Проверка статистики признаков**

````python
# список размерных признаков
length_columns = [
    'shell_length', 'shell_width',
    'head_length', 'head_width',
    'flipper_length_1', 'flipper_length_2',
    'flipper_length_3', 'flipper_length_4',
    'flipper_width_1', 'flipper_width_2',
    'flipper_width_3', 'flipper_width_4'
]
plt.figure(figsize=(14, 8))
sns.boxplot(data=df[length_columns])
plt.xticks(rotation=90)
plt.title('Boxplot размерных признаков')
plt.show()
df[length_columns].describe().T
````

По результатам describe() можно заметить, что максимальные значения некоторых признаков значительно превышают медиану и верхний квартиль. Такие значения в десятки раз превышают типичные размеры морских черепах, что подтверждает наличие аномалий и возможную ошибку масштабирования.

Вывод по статистике

Описательная статистика подтверждает результаты визуального анализа. Для некоторых признаков максимальные значения значительно превышают медиану и верхний квартиль.

Например:

shell_length: медиана ≈ 635 мм, максимум = 20240 мм

shell_width: медиана ≈ 487 мм, максимум = 11550 мм

flipper_length_1: медиана ≈ 416 мм, максимум = 2875 мм

Такие значения выглядят нереалистично и с высокой вероятностью связаны с ошибкой умножения измерений на 10.

**7.3 Исправление значений**

Чтобы убедиться в наличии аномалий, посмотрим самые большие значения для каждого признака.

````python
for col in length_columns:
    print(col, df[col].sort_values(ascending=False).head(10).values)
````

Этот шаг позволяет быстро увидеть, какие значения резко выделяются на фоне остальных. В некоторых признаках наблюдаются значения, значительно превышающие типичный диапазон измерений.Такие значения биологически нереалистичны для морских черепах и с высокой вероятностью возникли из-за ошибки умножения измерений на 10.

**7.4 Проверка нулевых значений**

````python
zero_counts = (df[length_columns] == 0).sum().sort_values(ascending=False)
zero_counts
````

Дополнительно были проверены нулевые значения размерных признаков. Для биометрических измерений нулевые значения выглядят нереалистично, поэтому такие наблюдения следует рассматривать как потенциальные ошибки записи или пропуски.

Если нулевые значения встречаются, их необходимо дополнительно обработать на этапе предобработки данных.

**7.5 Исправление значений**

На основе визуального анализа и статистики было принято решение исправить значения только в тех признаках, где наблюдаются аномально большие значения.

````python
anomaly_columns = [
    'shell_length',
    'shell_width',
    'flipper_length_1',
    'flipper_length_2',
    'flipper_length_3',
    'flipper_length_4'
]
df[anomaly_columns] = df[anomaly_columns].astype(float)
````

Исправление применялось только к признакам, в которых аномалии были подтверждены визуально и статистически. В качестве порога использовано значение 2000 мм, так как размеры, превышающие этот уровень, существенно выходят за реалистичный диапазон биометрических измерений черепах.

Деление таких значений на 10 позволяет скорректировать ошибку масштабирования и привести данные к единому формату.

````python
df.loc[df['shell_length'] > 3000, 'shell_length'] = df.loc[df['shell_length'] > 3000, 'shell_length'] / 10
df.loc[df['shell_width'] > 3000, 'shell_width'] = df.loc[df['shell_width'] > 3000, 'shell_width'] / 10

for col in ['flipper_length_1', 'flipper_length_2', 'flipper_length_3', 'flipper_length_4']:
    df.loc[df[col] > 2000, col] = df.loc[df[col] > 2000, col] / 10
````

Корректировка выполнялась только для тех признаков, в которых аномалии были подтверждены визуально и статистически.

Для `shell_length` и `shell_width` были исправлены значения выше 3000 мм. Для признаков `flipper_length_1–4` были исправлены значения выше 2000 мм. Такие значения существенно превышают реалистичный диапазон измерений и с высокой вероятностью являются результатом ошибочного умножения на 10.

Таким образом, исправление выполнено точечно: только для аномальных признаков и только для явно ошибочных значений.

На основе boxplot-диаграмм и описательной статистики были выявлены признаки, в которых присутствуют аномально большие значения: `shell_length`, `shell_width` и `flipper_length_1–4`.

Просмотр максимальных значений подтвердил, что часть наблюдений в этих столбцах многократно превышает основной диапазон данных и, вероятно, связана с ошибкой умножения измерений на 10.

Дополнительно были проверены нулевые значения размерных признаков.

Корректировка была выполнена точечно: только для аномальных признаков и только для значений, выходящих за реалистичный диапазон. Это позволяет привести данные к корректному масштабу и подготовить их к дальнейшему обучению модели.

## Масштабирование признаков

Масштабирование признаков необходимо для моделей, чувствительных к масштабу данных, таких как Ridge и Lasso.

В рамках проекта стандартизация будет реализована вручную. Для этого для каждого признака на обучающей выборке вычисляются:

среднее значение

стандартное отклонение

````python
# функция для стандартизации данных вручную
def standard_scaler_fit(X):
    mean_ = X.mean()
    std_ = X.std()
    std_ = std_.replace(0, 1)  # защита от деления на ноль
    return mean_, std_


def standard_scaler_transform(X, mean_, std_):
    return (X - mean_) / std_


# обучаем "скейлер" только на тренировочной выборке
train_mean, train_std = standard_scaler_fit(X_train)

# применяем масштабирование
X_train_scaled = standard_scaler_transform(X_train, train_mean, train_std)
X_val_scaled = standard_scaler_transform(X_val, train_mean, train_std)
X_test_scaled = standard_scaler_transform(X_test, train_mean, train_std)
````

**Вывод**

Масштабирование признаков было реализовано вручную без использования готовых функций библиотеки `sklearn`.

Параметры стандартизации — среднее значение и стандартное отклонение — были вычислены только на обучающей выборке. Затем эти параметры были применены к валидационной и тестовой выборкам.

Такой подход позволяет избежать утечки данных и корректно подготовить признаки для моделей, чувствительных к масштабу, таких как Ridge и Lasso.

## Функция расчёта метрик

Для оценки качества моделей регрессии создадим универсальную функцию, которая рассчитывает основные метрики качества.

Используем следующие метрики:

MAE (Mean Absolute Error) — средняя абсолютная ошибка предсказания

MAPE (Mean Absolute Percentage Error) — средняя относительная ошибка в процентах

R² (Coefficient of Determination) — доля объяснённой моделью дисперсии целевой переменной

Создание отдельной функции позволяет избежать дублирования кода и делает анализ моделей более структурированным.

````python
def calculate_metrics(y_true, y_pred):

    mae = mean_absolute_error(y_true, y_pred)

    mape = mean_absolute_percentage_error(y_true, y_pred)

    r2 = r2_score(y_true, y_pred)

    return {
        "MAE": mae,
        "MAPE": mape,
        "R2": r2
    }
````

## Сравнение моделей на train и validation

````python
def calculate_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    return {
        'MAE': round(mae, 2),
        'MAPE (%)': round(mape * 100, 2),
        'R2': round(r2, 4)
    }

models = {
    'DummyRegressor': dummy,
    'LinearRegression': lr,
    'Ridge': ridge,
    'Lasso': lasso
}

results = []

for name, model in models.items():
    train_pred = model.predict(X_train_scaled)
    val_pred = model.predict(X_val_scaled)

    train_metrics = calculate_metrics(y_train, train_pred)
    val_metrics = calculate_metrics(y_val, val_pred)

    results.append({
        'model': name,
        'train_MAE': train_metrics['MAE'],
        'val_MAE': val_metrics['MAE'],
        'train_MAPE (%)': train_metrics['MAPE (%)'],
        'val_MAPE (%)': val_metrics['MAPE (%)'],
        'train_R2': train_metrics['R2'],
        'val_R2': val_metrics['R2']
    })

results_df = pd.DataFrame(results)
results_df.sort_values(by='val_MAE')
````

## 20 Функция предсказания массы черепахи

После обучения и выбора лучшей модели можно создаем функцию, которая будет принимать измерения одной черепахи и возвращать предсказанную массу.

Функция будет:

принимать словарь с измерениями,

приводить данные к формату DataFrame,

масштабировать признаки,

делать предсказание с помощью обученной модели.

````python
def predict_turtle_weight(features_dict, model, scaler, feature_order):
    """
    Предсказывает массу черепахи на основе биометрических измерений.

    Параметры:
    features_dict : dict
        Словарь с измерениями черепахи.
        
    model :
        Обученная модель регрессии.
        
    scaler :
        Обученный StandardScaler.
        
    feature_order :
        Порядок признаков, использованный при обучении модели.

    Возвращает:
    float
        Предсказанная масса черепахи.
    """

    features_df = pd.DataFrame([features_dict])

    features_df = features_df[feature_order]

    features_scaled = scaler.transform(features_df)

    prediction = model.predict(features_scaled)

    return prediction[0]
````

<h2> Комментарий ревьюера  <a class="tocSkip"> </h2>
    
<b>Некоторые замечания и рекомендации💡:</b> 

Здесь совсем не то, что от нас требуется в задании.
    
Мы должны взять модель, вынуть из нее веса и смещение, а затем построить функцию, которая перемножает соответствующие признаки на веса, после чего добавляет смещение. Сумма произведение (скалярное произведение) весов и будет предсказанием модели. 
    
После чего мы можем взять предсказание моделью на одной какой-то строке, затем воспроизвести предсказание с помощью нашей функции, сравнить и убедиться, что получаем одно и то же.

```    
# Коэффициенты и смещение получаем так:
coef = best_model.coef_
intercept = best_model.intercept_
   
    

# Линейное предсказание: y = w1*x1 + w2*x2 + ... + w12*x12 + b
prediction = np.dot(coef, features) + intercept
return float(prediction)
```
