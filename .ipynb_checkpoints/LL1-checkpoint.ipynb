## Постановка задачи

Необходимо решить задачу регрессии: по набору морфологических признаков черепахи предсказать её массу.

**Целевая переменная:**
- `weight` — масса черепахи в килограммах.

**Основные требования к модели:**
- модель должна быть линейной;

**Дополнительная особенность данных:**
- часть измерений длины могла быть ошибочно увеличена в 10 раз, поэтому на этапе предобработки необходимо выявить и исправить подобные аномалии.

##  Исправление формата целевой переменной

Переводим weight в число.

Модель не сможет учиться, если целевая переменная хранится как текст.

````python
df['weight'] = pd.to_numeric(df['weight'].astype(str).str.replace(',', '.', regex=False), errors='coerce')
df['measure_count'] = pd.to_numeric(df['measure_count'].astype(str).str.replace(',', '.', regex=False), errors='coerce')
df['shell_crack'] = pd.to_numeric(df['shell_crack'].astype(str).str.replace(',', '.', regex=False), errors='coerce')

df['binomial_name'] = df['binomial_name'].astype(str).str.strip().str.lower()
````

9.3 Корреляции

````python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(14, 10))
sns.heatmap(numeric_df.corr(), cmap='coolwarm', center=0, annot=True, fmt='.2f')
plt.title('Корреляция числовых признаков')
plt.show()

````

Для оценки взаимосвязей между числовыми признаками была построена корреляционная матрица.

Анализ тепловой карты показывает наличие положительной корреляции между массой черепахи (`weight`) и её размерными характеристиками. Наиболее заметная связь наблюдается между массой и длиной/шириной панциря (`shell_length`, `shell_width`), а также длинами ласт (`flipper_length_*`). Это ожидаемый результат, так как более крупные черепахи, как правило, имеют большую массу.

Также наблюдается высокая корреляция между некоторыми размерными признаками между собой. Например, длина и ширина панциря, а также размеры ласт имеют сильную взаимосвязь. Это объясняется тем, что данные параметры описывают общие размеры тела черепахи.

Полученные результаты подтверждают, что биометрические параметры являются информативными признаками для построения модели линейной регрессии, поскольку они имеют значимую связь с целевой переменной — массой черепахи.

9.4 Boxplot размерных признаков

````python
length_columns = [
    'shell_length', 'shell_width',
    'head_length', 'head_width',
    'flipper_length_1', 'flipper_width_1',
    'flipper_length_2', 'flipper_width_2',
    'flipper_length_3', 'flipper_width_3',
    'flipper_length_4', 'flipper_width_4'
]

plt.figure(figsize=(14, 8))
sns.boxplot(data=df[length_columns])
plt.xticks(rotation=90)
plt.title('Boxplot размерных признаков')
plt.show()
````

Для оценки распределения биометрических параметров черепах был построен boxplot для всех размерных признаков.

На графике представлены распределения следующих признаков:

- размеры панциря (shell_length, shell_width)

- размеры головы (head_length, head_width)

- длина и ширина ласт (flipper_length_1–4, flipper_width_1–4)

Boxplot позволяет оценить медианные значения, межквартильный размах и наличие выбросов.

По графику можно заметить, что большинство измерений находятся в относительно компактных диапазонах значений. Например, размеры головы (head_length, head_width) имеют небольшой разброс и практически не содержат экстремальных значений.

В то же время для признаков, **связанных с размерами панциря (shell_length, shell_width), наблюдаются значительные выбросы, которые многократно превышают основной диапазон данных**. Аналогичные выбросы можно заметить и в некоторых признаках длины ласт (flipper_length_1, flipper_length_2, flipper_length_4).

Такие значения выглядят нереалистичными с биологической точки зрения и могут быть связаны с ошибками измерения. Согласно дополнительной информации от заказчика, **часть измерений могла быть ошибочно увеличена в 10 раз**.

Таким образом, анализ boxplot показывает наличие аномально больших значений в размерных признаках. На следующем этапе предобработки данных необходимо выявить такие наблюдения и скорректировать их (например, разделив значения на 10), чтобы избежать искажения модели машинного обучения.

Таким образом, выявленные значения подтверждают наличие аномалий в данных. На следующем этапе необходимо скорректировать такие наблюдения, разделив ошибочно увеличенные значения на 10.

**9.5 Исправление значений**

На основе визуального анализа и статистики было принято решение исправить значения только в тех признаках, где наблюдаются аномально большие значения.

````python
anomaly_columns = [
    'shell_length', 'shell_width',
    'head_length', 'head_width',
    'flipper_length_1', 'flipper_width_1',
    'flipper_length_2', 'flipper_width_2',
    'flipper_length_3', 'flipper_width_3',
    'flipper_length_4', 'flipper_width_4'
]
df[anomaly_columns] = df[anomaly_columns].astype(float)
````

Исправление применялось только к признакам, в которых аномалии были подтверждены визуально и статистически. В качестве порога использовано значение 2000 мм, так как размеры, превышающие этот уровень, существенно выходят за реалистичный диапазон биометрических измерений черепах.

Деление таких значений на 10 позволяет скорректировать ошибку масштабирования и привести данные к единому формату.

````python
df.loc[df['shell_length'] > 2000, 'shell_length'] = df.loc[df['shell_length'] > 2000, 'shell_length'] / 10
df.loc[df['shell_width'] > 1500, 'shell_width'] = df.loc[df['shell_width'] > 1500, 'shell_width'] / 10

df.loc[df['flipper_length_1'] > 1200, 'flipper_length_1'] = df.loc[df['flipper_length_1'] > 1200, 'flipper_length_1'] / 10
df.loc[df['flipper_length_2'] > 1200, 'flipper_length_2'] = df.loc[df['flipper_length_2'] > 1200, 'flipper_length_2'] / 10
df.loc[df['flipper_length_3'] > 1000, 'flipper_length_3'] = df.loc[df['flipper_length_3'] > 1000, 'flipper_length_3'] / 10
df.loc[df['flipper_length_4'] > 1000, 'flipper_length_4'] = df.loc[df['flipper_length_4'] > 1000, 'flipper_length_4'] / 10
````

В этом проекте аномалии корректируются только в тех признаках, где они действительно были выявлены при анализе графиков и статистики.
Используются пороговые значения, так как они позволяют точнее отделить нереалистичные наблюдения от реальных измерений.

Например:

для shell_length значения выше 2000 мм выглядят биологически неправдоподобно;

для shell_width значения выше 1500 мм также сильно выбиваются из основного диапазона;

для длин ласт пороги выбраны на основе статистики и максимальных значений.

Таким образом, корректируются только явно ошибочные наблюдения, а реалистичные значения сохраняются.

Мультиколлинеарность возникает, когда несколько признаков несут похожую информацию. В линейных моделях это может приводить к нестабильности коэффициентов и ухудшению интерпретируемости модели.

**9.9 Проверка мультиколлинеарности**

````python
X_num = df.drop(columns=['weight']).select_dtypes(include=['number']).copy()
X_num = X_num.fillna(X_num.median())

vif_data = pd.DataFrame()
vif_data["feature"] = X_num.columns
vif_data["VIF"] = [variance_inflation_factor(X_num.values, i) for i in range(X_num.shape[1])]

vif_data.sort_values("VIF", ascending=False)


````

timestamp — 162.6

Такие значения значительно превышают допустимые уровни и указывают на сильную мультиколлинеарность. Однако эти признаки не несут содержательной информации для прогнозирования массы черепахи:
id является техническим идентификатором записи, а timestamp отражает время измерения. Поэтому их следует рассматривать как служебные признаки и исключить из набора признаков перед обучением модели.

Во-вторых, ряд размерных признаков имеет умеренную мультиколлинеарность:

Такая зависимость объясняется тем, что все эти признаки описывают размеры тела черепахи и естественно коррелируют между собой. В биологических данных подобная взаимосвязь является ожидаемой.

В-третьих, признаки ширины ласт (flipper_width_1–4) имеют значения VIF около 5, что также указывает на умеренную взаимосвязь, но она не является критической.

## Масштабирование признаков

Масштабирование признаков необходимо для моделей, чувствительных к масштабу данных, таких как Ridge и Lasso.

В рамках проекта стандартизация будет реализована вручную. Для этого для каждого признака на обучающей выборке вычисляются:

среднее значение

стандартное отклонение

**16.1 Обучение без масштабирования**

Этот вариант используется как базовый уровень для сравнения качества моделей.

````python
X_train_no_scale = X_train.copy()
X_val_no_scale = X_val.copy()
X_test_no_scale = X_test.copy()
````

**16.2 Стандартизация признаков**

Стандартизация приводит признаки к распределению со средним 0 и стандартным отклонением 1.


````python
# функция для стандартизации данных вручную
def standard_scaler_fit(X):
    mean_ = X.mean()
    std_ = X.std()
    std_ = std_.replace(0, 1)
    return mean_, std_

def standard_scaler_transform(X, mean_, std_):
    return (X - mean_) / std_

# обучаем "скейлер" только на тренировочной выборке
train_mean, train_std = standard_scaler_fit(X_train)

X_train_scaled = standard_scaler_transform(X_train, train_mean, train_std)
X_val_scaled = standard_scaler_transform(X_val, train_mean, train_std)
X_test_scaled = standard_scaler_transform(X_test, train_mean, train_std)
````

6.3 Нормализация признаков (аналог MinMaxScaler)

MinMax-масштабирование приводит признаки к диапазону от 0 до 1.

````python
def minmax_scaler_fit(X):
    min_ = X.min()
    max_ = X.max()
    range_ = max_ - min_
    range_ = range_.replace(0, 1)  # защита от деления на ноль
    return min_, range_

def minmax_scaler_transform(X, min_, range_):
    return (X - min_) / range_

train_min, train_range = minmax_scaler_fit(X_train)

X_train_minmax = minmax_scaler_transform(X_train, train_min, train_range)
X_val_minmax = minmax_scaler_transform(X_val, train_min, train_range)
X_test_minmax = minmax_scaler_transform(X_test, train_min, train_range)
````

В рамках подготовки данных были реализованы три варианта масштабирования признаков:

без масштабирования — используется как базовый вариант для сравнения моделей;

стандартизация признаков — приводит признаки к распределению со средним 0 и стандартным отклонением 1;

MinMax-нормализация — масштабирует признаки в диапазон от 0 до 1.

Все методы масштабирования реализованы вручную без использования готовых функций библиотеки sklearn.
Параметры масштабирования вычислялись только на тренировочной выборке, после чего применялись к валидационной и тестовой выборкам, что позволяет избежать утечки информации и корректно оценить качество моделей. Такой подход позволяет избежать утечки данных и корректно подготовить признаки для моделей, чувствительных к масштабу, таких как Ridge и Lasso.

## Функция расчёта метрик

Для оценки качества моделей регрессии создадим универсальную функцию, которая будет рассчитывать основные метрики качества предсказаний.

В рамках проекта используются следующие метрики:

**MAE (Mean Absolute Error)** — средняя абсолютная ошибка предсказаний.  
Показывает, на сколько килограммов в среднем модель ошибается при прогнозировании массы черепах.

**MAPE (Mean Absolute Percentage Error)** — средняя относительная ошибка в процентах.  
Данная метрика позволяет оценить ошибку модели с учётом масштаба значений, что особенно важно при прогнозировании массы черепах разного размера.

**R² (Coefficient of Determination)** — коэффициент детерминации, который показывает долю вариации целевой переменной, объясняемую моделью.

Для корректного расчёта MAPE используется отдельная функция `safe_mape`, которая исключает возможное деление на ноль. Это делает расчёт метрики более устойчивым.

Создание отдельной функции для расчёта метрик позволяет избежать дублирования кода и делает процесс оценки моделей более удобным и структурированным.

````python
def safe_mape(y_true, y_pred):
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

def calculate_metrics(y_true, y_pred):
    return {
        'MAE': round(mean_absolute_error(y_true, y_pred), 2),
        'MAPE (%)': round(safe_mape(y_true, y_pred), 2),
        'R2': round(r2_score(y_true, y_pred), 4)
    }
````

## Сравнение моделей на train и validation

````python
models = {
    'DummyRegressor': dummy,
    'LinearRegression': lr,
    'Ridge': ridge,
    'Lasso': lasso
}

results = []

for name, model in models.items():
    train_pred = model.predict(X_train_scaled)
    val_pred = model.predict(X_val_scaled)

    train_metrics = calculate_metrics(y_train, train_pred)
    val_metrics = calculate_metrics(y_val, val_pred)

    results.append({
        'model': name,
        'train_MAE': train_metrics['MAE'],
        'val_MAE': val_metrics['MAE'],
        'train_MAPE (%)': train_metrics['MAPE (%)'],
        'val_MAPE (%)': val_metrics['MAPE (%)'],
        'train_R2': train_metrics['R2'],
        'val_R2': val_metrics['R2']
    })

results_df = pd.DataFrame(results)
results_df.sort_values(by='val_R2', ascending=False)
````

## Выбор лучшей модели

Подберем гиперпарметры лучшей модели 

**Подбор параметров Ridge**

````python
alphas = [0.001, 0.01, 0.1, 1, 10, 100]

ridge_results = []

for alpha in alphas:
    
    model = Ridge(alpha=alpha)
    model.fit(X_train_scaled, y_train)
    
    pred = model.predict(X_val_scaled)
    
    metrics = calculate_metrics(y_val, pred)
    
    ridge_results.append({
        'alpha': alpha,
        'MAE': metrics['MAE'],
        'R2': metrics['R2']
    })

ridge_results_df = pd.DataFrame(ridge_results)
ridge_results_df
````

````python
alphas = [0.0001, 0.001, 0.01, 0.1, 1]

lasso_results = []

for alpha in alphas:
    
    model = Lasso(alpha=alpha, max_iter=20000)
    model.fit(X_train_scaled, y_train)
    
    pred = model.predict(X_val_scaled)
    
    metrics = calculate_metrics(y_val, pred)
    
    lasso_results.append({
        'alpha': alpha,
        'MAE': metrics['MAE'],
        'R2': metrics['R2']
    })

lasso_results_df = pd.DataFrame(lasso_results)
lasso_results_df
````

Аналогичным образом был выполнен подбор гиперпараметра alpha для модели Lasso.
Для каждого значения параметра модель обучалась на тренировочной выборке и оценивалась на валидационной. Это позволяет определить оптимальную силу регуляризации для данной модели.

После этого выбираем лучший alpha

````python
best_ridge_alpha = ridge_results_df.sort_values(by='R2', ascending=False).iloc[0]['alpha']
best_lasso_alpha = lasso_results_df.sort_values(by='R2', ascending=False).iloc[0]['alpha']

print("Лучший alpha Ridge:", best_ridge_alpha)
print("Лучший alpha Lasso:", best_lasso_alpha)
````

После расчёта метрик для разных значений параметра alpha было выбрано значение, при котором модель показывает наилучшее качество по метрике R². Это значение используется для обучения финальной версии модели.

Обучаем финальные модели

````python
ridge = Ridge(alpha=best_ridge_alpha)
ridge.fit(X_train_scaled, y_train)

lasso = Lasso(alpha=best_lasso_alpha, max_iter=20000)
lasso.fit(X_train_scaled, y_train)
````

После подбора оптимальных гиперпараметров были обучены финальные версии моделей Ridge и Lasso с выбранными значениями alpha. Эти модели будут использованы для дальнейшего сравнения с LinearRegression и выбора лучшей модели для решения задачи прогнозирования массы черепах.

**Вывод по подбору гиперпараметров**

Для моделей Ridge и Lasso был выполнен подбор гиперпараметра регуляризации alpha.
Для этого были протестированы различные значения параметра и рассчитаны метрики качества на валидационной выборке.

Результаты показали, что изменение параметра регуляризации практически не влияет на качество модели. Значения MAE и R² остаются близкими для различных значений alpha.

Это говорит о том, что данные хорошо описываются линейной зависимостью и регуляризация не даёт значительного улучшения.

##  Функция предсказания массы черепахи

После обучения и выбора лучшей модели можно создаем функцию, которая будет принимать измерения одной черепахи и возвращать предсказанную массу.

Функция будет:

принимать словарь с измерениями,

приводить данные к формату DataFrame,

масштабировать признаки,

делать предсказание с помощью обученной модели.

**Получение коэффициентов и смещения**

````python
coef = best_model.coef_
intercept = best_model.intercept_
````

**Функция ручного предсказания**

````python

def linear_predict(features, coef, intercept):
    prediction = np.dot(coef, features) + intercept
    return float(prediction)
````

**Проверка работоспособности функции**

Запуск предсказания

````python
sample_features = X_test_scaled.iloc[0].values

model_prediction = best_model.predict([sample_features])[0]
manual_prediction = linear_predict(sample_features, coef, intercept)

print("Предсказание модели:", round(model_prediction, 6))
print("Предсказание вручную:", round(manual_prediction, 6))
````

**Вывод**

Для лучшей модели была реализована функция ручного предсказания по формуле линейной регрессии.
Сравнение показало, что предсказание, полученное с помощью функции, совпадает с результатом метода predict() модели. Это подтверждает корректность реализации.

Результаты совпали:

- Предсказание модели: 86.812533
- Предсказание вручную: 86.812533


Совпадение значений подтверждает, что предсказание линейной модели действительно вычисляется как скалярное произведение признаков и коэффициентов с добавлением свободного члена.
