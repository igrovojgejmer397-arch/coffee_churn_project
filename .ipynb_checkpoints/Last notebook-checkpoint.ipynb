Модель не сможет учиться, если целевая переменная хранится как текст.

## Поиск и исправление ошибок умножениея на ×10

В дополнительном сообщении заказчик сообщил, что при внесении данных некоторые измерения длины могли быть ошибочно умножены на 10. Это произошло из-за того, что один из волонтёров предположил, что значения записаны в сантиметрах, и дополнительно умножил их на 10, хотя все измерения должны быть указаны в миллиметрах.

Такие ошибки могут привести к появлению аномально больших значений, которые будут искажать статистику признаков и ухудшать качество модели линейной регрессии. Поэтому необходимо проверить размерные признаки и при необходимости скорректировать такие значения.

**7.1 Определение размерных признаков**

На этапе исследовательского анализа данных ранее были построены boxplot-диаграммы для размерных признаков. Эти графики позволили визуально выявить наличие аномально больших значений.

Наиболее заметные выбросы наблюдаются в признаках:

shell_length

shell_width

flipper_length_1

flipper_length_2

flipper_length_3

flipper_length_4

В этих признаках присутствуют значения, значительно превышающие основной диапазон наблюдений. Это может свидетельствовать о возможной ошибке масштабирования (умножении значений на 10).

Таким образом, визуальный анализ указывает на необходимость дополнительной проверки размерных признаков.

**7.2 Проверка статистики признаков**

````python
length_columns = [
    'shell_length', 'shell_width',
    'head_length', 'head_width',
    'flipper_length_1', 'flipper_width_1',
    'flipper_length_2', 'flipper_width_2',
    'flipper_length_3', 'flipper_width_3',
    'flipper_length_4', 'flipper_width_4'
]

for col in length_columns:
    median_val = df[col].median()
    mask = df[col] > median_val * 4
    df.loc[mask, col] = df.loc[mask, col] / 10
````

````python
df[length_columns].describe().T
````

## Масштабирование признаков

Масштабирование признаков необходимо для моделей, чувствительных к масштабу данных, таких как Ridge и Lasso.

В рамках проекта стандартизация будет реализована вручную. Для этого для каждого признака на обучающей выборке вычисляются:

среднее значение

стандартное отклонение

````python
# функция для стандартизации данных вручную
def standard_scaler_fit(X):
    mean_ = X.mean()
    std_ = X.std()
    std_ = std_.replace(0, 1)
    return mean_, std_

def standard_scaler_transform(X, mean_, std_):
    return (X - mean_) / std_

# обучаем "скейлер" только на тренировочной выборке
train_mean, train_std = standard_scaler_fit(X_train)

X_train_scaled = standard_scaler_transform(X_train, train_mean, train_std)
X_val_scaled = standard_scaler_transform(X_val, train_mean, train_std)
X_test_scaled = standard_scaler_transform(X_test, train_mean, train_std)
````

**Вывод**

Масштабирование признаков было реализовано вручную без использования готовых функций библиотеки `sklearn`.

Параметры стандартизации — среднее значение и стандартное отклонение — были вычислены только на обучающей выборке. Затем эти параметры были применены к валидационной и тестовой выборкам.

Такой подход позволяет избежать утечки данных и корректно подготовить признаки для моделей, чувствительных к масштабу, таких как Ridge и Lasso.

## Функция расчёта метрик

Для оценки качества моделей регрессии создадим универсальную функцию, которая будет рассчитывать основные метрики качества предсказаний.

В рамках проекта используются следующие метрики:

**MAE (Mean Absolute Error)** — средняя абсолютная ошибка предсказаний.  
Показывает, на сколько килограммов в среднем модель ошибается при прогнозировании массы черепах.

**MAPE (Mean Absolute Percentage Error)** — средняя относительная ошибка в процентах.  
Данная метрика позволяет оценить ошибку модели с учётом масштаба значений, что особенно важно при прогнозировании массы черепах разного размера.

**R² (Coefficient of Determination)** — коэффициент детерминации, который показывает долю вариации целевой переменной, объясняемую моделью.

Для корректного расчёта MAPE используется отдельная функция `safe_mape`, которая исключает возможное деление на ноль. Это делает расчёт метрики более устойчивым.

Создание отдельной функции для расчёта метрик позволяет избежать дублирования кода и делает процесс оценки моделей более удобным и структурированным.

````python
def safe_mape(y_true, y_pred):
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

def calculate_metrics(y_true, y_pred):
    return {
        'MAE': round(mean_absolute_error(y_true, y_pred), 2),
        'MAPE (%)': round(safe_mape(y_true, y_pred), 2),
        'R2': round(r2_score(y_true, y_pred), 4)
    }
````

## Обучение моделей

````python
dummy = DummyRegressor(strategy='mean')
dummy.fit(X_train_scaled, y_train)

lr = LinearRegression()
lr.fit(X_train_scaled, y_train)

ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, y_train)

lasso = Lasso(alpha=0.01, max_iter=20000)
lasso.fit(X_train_scaled, y_train)
````

**Вывод**

На данном этапе были обучены четыре модели:

- DummyRegressor — базовая модель, которая всегда предсказывает среднее значение целевой переменной;
- LinearRegression;
- Ridge;
- Lasso.

Использование dummy-модели позволяет оценить, дают ли линейные модели реальное улучшение качества по сравнению с наивным базовым подходом.

Далее модели будут сравнены на валидационной выборке по метрикам качества.

## Сравнение моделей на train и validation

````python
models = {
    'DummyRegressor': dummy,
    'LinearRegression': lr,
    'Ridge': ridge,
    'Lasso': lasso
}

results = []

for name, model in models.items():
    train_pred = model.predict(X_train_scaled)
    val_pred = model.predict(X_val_scaled)

    train_metrics = calculate_metrics(y_train, train_pred)
    val_metrics = calculate_metrics(y_val, val_pred)

    results.append({
        'model': name,
        'train_MAE': train_metrics['MAE'],
        'val_MAE': val_metrics['MAE'],
        'train_MAPE (%)': train_metrics['MAPE (%)'],
        'val_MAPE (%)': val_metrics['MAPE (%)'],
        'train_R2': train_metrics['R2'],
        'val_R2': val_metrics['R2']
    })

results_df = pd.DataFrame(results)
results_df.sort_values(by='val_R2', ascending=False)
````

##  Функция предсказания массы черепахи

После обучения и выбора лучшей модели можно создаем функцию, которая будет принимать измерения одной черепахи и возвращать предсказанную массу.

Функция будет:

принимать словарь с измерениями,

приводить данные к формату DataFrame,

масштабировать признаки,

делать предсказание с помощью обученной модели.

````python
def predict_turtle_weight(sample, model, mean_, std_, feature_columns):
    sample_df = pd.DataFrame([sample])
    sample_df = sample_df[feature_columns]
    sample_scaled = (sample_df - mean_) / std_
    prediction = model.predict(sample_scaled)
    return prediction[0]
````

<div class="alert alert-warning">
<h2> Комментарий ревьюера  <a class="tocSkip"> </h2>
    
<b>Некоторые замечания и рекомендации💡:</b>

Мы должны взять модель, вынуть из нее веса и смещение, а затем построить функцию, которая перемножает соответствующие признаки на веса, после чего добавляет смещение. Сумма произведение (скалярное произведение) весов и будет предсказанием модели. 
    
После чего мы можем взять предсказание моделью на одной какой-то строке, затем воспроизвести предсказание с помощью нашей функции, сравнить и убедиться, что получаем одно и то же.

```    
# Коэффициенты и смещение получаем так:
coef = best_model.coef_
intercept = best_model.intercept_
   
    

# Линейное предсказание: y = w1*x1 + w2*x2 + ... + w12*x12 + b
prediction = np.dot(coef, features) + intercept
return float(prediction)
```
