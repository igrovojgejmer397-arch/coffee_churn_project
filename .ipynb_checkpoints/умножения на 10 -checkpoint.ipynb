##  Исправление формата целевой переменной

Переводим weight в число.

Модель не сможет учиться, если целевая переменная хранится как текст.

````python
df['weight'] = pd.to_numeric(df['weight'].astype(str).str.replace(',', '.', regex=False), errors='coerce')
df['measure_count'] = pd.to_numeric(df['measure_count'].astype(str).str.replace(',', '.', regex=False), errors='coerce')
df['shell_crack'] = pd.to_numeric(df['shell_crack'].astype(str).str.replace(',', '.', regex=False), errors='coerce')

df['binomial_name'] = df['binomial_name'].astype(str).str.strip().str.lower()
````

Это ожидаемый результат, так как более крупные черепахи, как правило, имеют большую массу.

## Поиск и исправление ошибок умножениея на ×10

В дополнительном сообщении заказчик сообщил, что при внесении данных некоторые измерения длины могли быть ошибочно умножены на 10. Это произошло из-за того, что один из волонтёров предположил, что значения записаны в сантиметрах, и дополнительно умножил их на 10, хотя все измерения должны быть указаны в миллиметрах.

Такие ошибки могут привести к появлению аномально больших значений, которые будут искажать статистику признаков и ухудшать качество модели линейной регрессии. Поэтому необходимо проверить размерные признаки и при необходимости скорректировать такие значения.

**7.1 Определение размерных признаков**

На этапе исследовательского анализа данных ранее были построены boxplot-диаграммы для размерных признаков. Эти графики позволили визуально выявить наличие аномально больших значений.

Наиболее заметные выбросы наблюдаются в признаках:

shell_length

shell_width

flipper_length_1

flipper_length_2

flipper_length_3

flipper_length_4

В этих признаках присутствуют значения, значительно превышающие основной диапазон наблюдений. Это может свидетельствовать о возможной ошибке масштабирования (умножении значений на 10).

Таким образом, визуальный анализ указывает на необходимость дополнительной проверки размерных признаков.

**7.2 Проверка статистики признаков**

````python
length_columns = [
    'shell_length', 'shell_width',
    'head_length', 'head_width',
    'flipper_length_1', 'flipper_width_1',
    'flipper_length_2', 'flipper_width_2',
    'flipper_length_3', 'flipper_width_3',
    'flipper_length_4', 'flipper_width_4'
]

for col in length_columns:
    median_val = df[col].median()
    mask = df[col] > median_val * 4
    df.loc[mask, col] = df.loc[mask, col] / 10
````

````python
df[length_columns].describe().T
````

По результатам describe() можно заметить, что максимальные значения некоторых признаков значительно превышают медиану и верхний квартиль. Такие значения в десятки раз превышают типичные размеры морских черепах, что подтверждает наличие аномалий и возможную ошибку масштабирования.

Вывод по статистике

Описательная статистика подтверждает результаты визуального анализа. Для некоторых признаков максимальные значения значительно превышают медиану и верхний квартиль.

Например:

shell_length: медиана ≈ 635 мм, максимум = 20240 мм

shell_width: медиана ≈ 487 мм, максимум = 11550 мм

flipper_length_1: медиана ≈ 416 мм, максимум = 2875 мм

Такие значения выглядят нереалистично и с высокой вероятностью связаны с ошибкой умножения измерений на 10.

**7.3 Исправление  максимальных значений**

Чтобы убедиться в наличии аномалий, посмотрим самые большие значения для каждого признака.

````python
for col in length_columns:
    print(col, df[col].sort_values(ascending=False).head(10).values)
````

Этот шаг позволяет быстро увидеть, какие значения резко выделяются на фоне остальных. В некоторых признаках наблюдаются значения, значительно превышающие типичный диапазон измерений.Такие значения биологически нереалистичны для морских черепах и с высокой вероятностью возникли из-за ошибки умножения измерений на 10.

**7.4 Проверка нулевых значений**

````python
zero_counts = (df[length_columns] == 0).sum().sort_values(ascending=False)
zero_counts
````

Дополнительно были проверены нулевые значения размерных признаков. Для биометрических измерений нулевые значения выглядят нереалистично, поэтому такие наблюдения следует рассматривать как потенциальные ошибки записи или пропуски.

Если нулевые значения встречаются, их необходимо дополнительно обработать на этапе предобработки данных.

**7.5 Исправление значений**

На основе визуального анализа и статистики было принято решение исправить значения только в тех признаках, где наблюдаются аномально большие значения.

````python
anomaly_columns = [
    'shell_length',
    'shell_width',
    'flipper_length_1',
    'flipper_length_2',
    'flipper_length_3',
    'flipper_length_4'
]
df[anomaly_columns] = df[anomaly_columns].astype(float)
````

Исправление применялось только к признакам, в которых аномалии были подтверждены визуально и статистически. В качестве порога использовано значение 2000 мм, так как размеры, превышающие этот уровень, существенно выходят за реалистичный диапазон биометрических измерений черепах.

Деление таких значений на 10 позволяет скорректировать ошибку масштабирования и привести данные к единому формату.

````python
df.loc[df['shell_length'] > 3000, 'shell_length'] = df.loc[df['shell_length'] > 3000, 'shell_length'] / 10
df.loc[df['shell_width'] > 3000, 'shell_width'] = df.loc[df['shell_width'] > 3000, 'shell_width'] / 10

for col in ['flipper_length_1', 'flipper_length_2', 'flipper_length_3', 'flipper_length_4']:
    df.loc[df[col] > 2000, col] = df.loc[df[col] > 2000, col] / 10
````

Корректировка выполнялась только для тех признаков, в которых аномалии были подтверждены визуально и статистически.

Для `shell_length` и `shell_width` были исправлены значения выше 3000 мм. Для признаков `flipper_length_1–4` были исправлены значения выше 2000 мм. Такие значения существенно превышают реалистичный диапазон измерений и с высокой вероятностью являются результатом ошибочного умножения на 10.

Таким образом, исправление выполнено точечно: только для аномальных признаков и только для явно ошибочных значений.

На основе boxplot-диаграмм и описательной статистики были выявлены признаки, в которых присутствуют аномально большие значения: `shell_length`, `shell_width` и `flipper_length_1–4`.

Просмотр максимальных значений подтвердил, что часть наблюдений в этих столбцах многократно превышает основной диапазон данных и, вероятно, связана с ошибкой умножения измерений на 10.

Дополнительно были проверены нулевые значения размерных признаков.

Корректировка была выполнена точечно: только для аномальных признаков и только для значений, выходящих за реалистичный диапазон. Это позволяет привести данные к корректному масштабу и подготовить их к дальнейшему обучению модели.
